```python
%matplotlib inline
import time
import pylab as pl
from IPython import display
```

```python
import serial # Import Serial Library
import numpy # Import numpy
import matplotlib.pyplot as plt # Import matplotlib library
from drawnow import *
```

```python
arduinoData = serial.Serial('/dev/cu.usbserial-1410', 9600) # Create serial object named arduinoData

plt.ion() # Tell matplotlib you want interactive mode to plot live data

cnt=0
```

```python
TempThreshold = []
tempC= []

while True: # While loop that loops forever
    while (arduinoData.inWaiting()==0): # Wait here until there is data
        pass # Do nothing
                    # Plot the legend
    arduinoString = arduinoData.readline() # Read the line of text from the serial port
    dataArray = arduinoString.split(',')   # Split it into an array called dataArray
    temp = float(dataArray[0])             # Convert first element to floating number and put in temp
    tempC.append(temp)                     # Build our tempC array by appending temp readings
    TempThreshold.append(23)
    plt.figure(figsize=(17, 6))
    pl.plot(TempThreshold, 'k--')
    pl.plot(tempC, 'r-', label='Degrees F')
    pl.ylim(20, 25)
    pl.xlabel('Time')
    pl.ylabel('Temperature (C)')
    display.clear_output(wait=True)
    display.display(pl.gcf())
    #time.sleep(1.0)
    plt.pause(.0001)
```
